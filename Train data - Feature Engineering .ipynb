## We will be performing all the below steps in Feature Engineering :-

1. Handling Missing Values
2. Temporal variables ( eg - Datetime vaiables ) 
3. Categorical Variables : Remove rare labels
4.  Replace categorical features with numerical values
5. Standardize the value of the variables to the same range  
    
    
    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# to visualize all the columns of the dataset
pd.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("price_train.csv")
dataset.head()
```

### Handling Missing Values

```python
# First we handle categorical features with missing values

feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes == 'O']

for feature in feature_nan:
    print(feature," : {} % of missing value".format(np.round(dataset[feature].isnull().mean(),4)))
    
```

```python
# Replacing the missing values with new label
for feature in feature_nan:
    dataset[feature] = dataset[feature].fillna('Missing')

dataset[feature_nan].isnull().sum()    
    
```

```python
dataset.head()
```

```python
# Now we handle the missing values in numerical features

# check the numerical variables which have the missing values

numerical_with_nan = [ feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes != 'O']

# lets print the % of missing values in each  numerical feature
for feature in numerical_with_nan:
    print(feature," has {} % of missing value".format(np.round(dataset[feature].isnull().mean(),4)))
```

```python
# Replacing the Missig values

for feature in numerical_with_nan:
    feature_median = dataset[feature].median()
    
    # we create new features to capture nan values 
    dataset[feature+'_nan'] = np.where(dataset[feature].isnull(),1,0)
    # we replace the missing values with median because outliers is present
    dataset[feature] = dataset[feature].fillna(feature_median)

dataset[numerical_with_nan].isnull().sum()    
```

```python
dataset.head(20)
```

```python
## Temporal variables ( Date Time variables)

# we capture the years differnce values with respect to year of sold 

for feature in ["YearBuilt","YearRemodAdd","GarageYrBlt"]:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset[["YearBuilt","YearRemodAdd","GarageYrBlt"]].head()
```

```python
dataset.head()
```

#### now we will do log transformation to those continoues numerical feature which are not following normal distribution( we saw in data analysis)


```python

num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

## Handling the Rare Categorical Feature

we will remove the categorical variables that are present less then 1 % of the observation and replace them with new label

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == "O"]
```

```python
for feature in categorical_features:
    print(dataset[feature].value_counts())
```

```python
# now handle the rare categories present ( less the 1 % ) in categorical features and replace the wtih new lable

for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature]= np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
    
```

```python
dataset.head(40)
```

#### Replacing categorical features with numerical values

we are using frequent odinal Encoding

```python
for feature in categorical_features:
    label_ordered = dataset.groupby(feature)['SalePrice'].count().sort_values().to_dict()
    dataset[feature] = dataset[feature].map(label_ordered)
   
```

```python
dataset.head()
```

# Feature Scaling

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
feature_scale
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
 scaler.transform(dataset[feature_scale])
```

```python
# scaling the test data

test_data = pd.read_csv("test_feature_eng.csv")
test_data = test_data.drop(columns =['Id'],axis = 1)

test_data = scaler.transform(test_data[feature_scale])
test_data = pd.DataFrame(test_data,columns = feature_scale)
test_data.head()
```

```python
test_data.to_csv('test_scaled.csv',index = False)
```

```python
data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop = True),
                      pd.DataFrame(scaler.transform(dataset[feature_scale]), columns = feature_scale)],axis = 1)
```
