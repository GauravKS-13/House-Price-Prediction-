- Missing Values

### Missing Values

```python
# check the nan value percentage in each feature

# Step 1 -  list of features which have missing values
feature_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1]

# Step 2 - printing the feature name and missing value(nan value) percentage
for feature in feature_with_na:
    print(feature,"-" , np.round(dataset[feature].isnull().mean(), 4),"% missing values")
```

### Since there are many missing values ,we need to find the relationship between missing values and Sales Price

- we will plot some diagrams for the relationship

```python
for feature in feature_with_na:
    # making a copy of data set,so our original data does not get effected.
    data = dataset.copy() 
    
    # Let's make a variable that indicates 1 if nan value present otherwise 0
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    # Lets make a plot for the median salePrice where value is missing and present
    data.groupby(feature)['SalePrice'].median().plot.bar(color = ['red','blue'])
    plt.title(feature)
    plt.show()
    
    
```

#### Here,We can see that there is some relation between missing values and SalePrice, so we have to replace the nan values(missing values) with something meaningful, this we will do in Feature Engineering.


we can see that feature Id is not required so we have to drop this in feature Engineering

### Numerical Features/Variables

```python
# Listing the numerical features
numerical_features = [ feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

# total numerical feature
print("Total numerical features is", len(numerical_features))

# Looking at all the numerical features
dataset[numerical_features].head()
```

#### Temporal Variable ( Eg - Datetime Variables )

Here In this Dataset, we have 4 year variables. one specific scenario can be ,difference of years between house is build and house is sold.

```python
# List of features that contains year information
year_features = [feature for feature in numerical_features if 'Year' in feature or 'Yr' in feature]
year_features
```

```python
# Explore the year featurs

for feature in year_features:
    print(feature,"-",dataset[feature].unique())
    print()
```

```python
# Let's analyze the temporal variables

# we will check there is a relationship between year sold and sales price
plt.figure(figsize=(8,6))
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Yrsold')
plt.ylabel("Sale Price")
plt.show()
```

####  
here we see that saleprice is decreasing with year of sold increasing

```python
# Now we will compare the difference between all year feature with SalePrice

for feature in year_features:
    if feature != 'YrSold':
        data = dataset.copy()
         # we will capture the difference between year feature and year of house sold
        data[feature] = data['YrSold'] - data[feature]
    
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

#### Numerical Variables are two type
1. Continuous ariable
2. Discrete Variable


```python
# make the list of discrete features
discrete_features = [ feature  for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_features + ['Id']]
print("total discrete feature is",len(discrete_features))
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

```python
# Let's find the relationship between discrete features ans Sale Price (Dependent Feature)

for feature in discrete_features:
    dataset.groupby(feature)['SalePrice'].median().plot.bar(color = ['red','green','blue','orange','black'])
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
    
```

  Here we can see there is some relation between discrete variavles and Sale Price

```python
# List of continuous Variables/features
continuous_features = [ feature for feature in numerical_features if feature not in discrete_features and feature not in year_features+['Id']]
print("total continuous features is ",len(continuous_features))
```

```python
continuous_features
```

```python
# Let's analyze the continuous variables by creating the Histogram to understand the distribution of each continuous feature

for feature in continuous_features:
    data = dataset.copy()
    data[feature].hist(bins = 25 , color = 'red')
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
    
```

```python
## we will be using log transformation   

for feature in continuous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
        
```

### Outliers

```python
for feature in continuous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column= feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

### Categorical Variables

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
print(len(categorical_features))
categorical_features
```

```python
dataset[categorical_features].head()
```

```python
### check how many unique categorices each feature have

for feature in categorical_features:
    print("The feature is {} and nuber of categories are {}".format(feature,len(dataset[feature].unique())))
```

```python
# finding the relationship between categorical features and  Sale price

for feature in categorical_features:
    dataset.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Sale Price")
    plt.title(feature)
    plt.show()
```

In this visualization we can see that which category has the highest meadian salary

```python
for feature in categorical_features:
    sns.countplot(x = feature , data = dataset)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```
