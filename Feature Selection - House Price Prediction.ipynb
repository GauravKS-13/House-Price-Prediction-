### Feature Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("feature_train.csv")
dataset.head()
```

```python
# dependent feature
y_train = dataset[['SalePrice']]
y_train.shape
```

```python
# drop the dependent feature and unrequired columns and get Independent features
x_train = dataset.drop(columns = ['Id','SalePrice'], axis = 1)
x_train.shape
```

```python
#  for Feature Selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
### Apply the feature selection
# first specify the lasso regression model than
# chooose the suitable value of alpha ( equivalent of penlety)
# higher the alpha value less featue will be selected

# then we use the SelectFromModle from sklearn which will select the features which coefficient are non zero

feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0)) # Remember to set the seed and random_state same will use in test data
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

#### Feature those have True value they are selected and have non-zero coefficient value and features which has False are not Selected and have zero coefficient value

```python
# This is how we can make the list of selected features
selected_features = x_train.columns[(feature_sel_model.get_support())]
print(selected_features,"\n")


# let's print the total number of feature ad seected features

print("selected features are : {}".format(len(selected_features)))
print("total feature are :{}".format(len(x_train.columns)))
print("feature with coefficient shrunk to zero are :{}".format(82-21))
```

#### we get 21 features are selected and we will use these feature to create our model
Note - for better feature selection you should always do better feature engineering

```python
x_train = x_train[selected_features]
y_train
```
